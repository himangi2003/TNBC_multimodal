```python
import os
import numpy as np
import cv2
from tqdm import tqdm
from PIL import Image
import pandas as pd
from tiatoolbox.wsicore.wsireader import WSIReader, VirtualWSIReader
from tiatoolbox.tools.tissuemask import OtsuTissueMasker
from tiatoolbox.tools.patchextraction import SlidingWindowPatchExtractor


# === Configuration ===
data_path = '/lab/deasylab3/Jung/tiger/'
dir_TIFF_images = os.path.join(data_path, "wsirois/wsi-level-annotations/images/")
out_tiles_dir = "tiles_pipeline2"
patch_size = 224
stride = 112
min_mask_ratio = 0.5


# === Helper: Crop low-res region ===
def crop_bounding_box_from_wsi(wsi_path, bounds, lowres_level=None):
    reader = WSIReader.open(wsi_path)
    if lowres_level is None:
        lowres_level = reader.info.level_count - 1

    crop_region = reader.read_bounds(bounds, resolution=lowres_level, units="level")
    downsample = reader.info.level_downsamples[lowres_level]
    
    return crop_region, reader, downsample


# === Helper: Generate Otsu tissue mask ===
def generate_tissue_mask_otsu_tiatoolbox(thumbnail, bounds, full_dims, downsample):
    masker = OtsuTissueMasker()
    masker.fit([thumbnail])
    masks = masker.transform([thumbnail])
    lowres_mask = masks[0].astype("uint8") * 255

    h = bounds[3] - bounds[1]
    w = bounds[2] - bounds[0]
    fullres_mask = cv2.resize(lowres_mask, (w, h), interpolation=cv2.INTER_NEAREST)

    canvas = np.zeros(full_dims[::-1], dtype="uint8")
    canvas[bounds[1]:bounds[3], bounds[0]:bounds[2]] = fullres_mask

    return canvas


# === Helper: Tile and save patches ===
def tile_and_save_patches(reader, mask_array, out_dir, slide_id,
                          patch_size=224, stride=112, min_mask_ratio=0.5):
    os.makedirs(out_dir, exist_ok=True)
        # Ensure proper binary mask format for VirtualWSIReader
    # Binarize to 0 and 255
    mask_array = ((mask_array > 127).astype(np.uint8)) * 255
    
    # Ensure it's (H, W, 1), dtype=uint8
    if mask_array.ndim == 2:
        mask_array = np.expand_dims(mask_array, axis=-1)
    
    assert mask_array.dtype == np.uint8, "Mask must be uint8"
    assert set(np.unique(mask_array.flatten())) <= {0, 255}, "Mask must be binary"
    
    # Feed to VirtualWSIReader
    mask_reader = VirtualWSIReader(mask_array)



    extractor = SlidingWindowPatchExtractor(
        input_img=reader,
        patch_size=patch_size,
        stride=stride,
        resolution=0,
        units="level",
        input_mask=mask_reader,
        within_bound=True,
        min_mask_ratio=min_mask_ratio,
    )

    coords = extractor.locations_df[["x", "y"]]
    metadata = []

    for i, patch in enumerate(tqdm(extractor, desc=f"Tiling {slide_id}")):
        x, y = int(coords.iloc[i]["x"]), int(coords.iloc[i]["y"])
        filename = f"{slide_id}_tile_{i}_x{x}_y{y}.png"
        Image.fromarray(patch).save(os.path.join(out_dir, filename))
        metadata.append({
            "index": i, "filename": filename, "x": x, "y": y, "slide": slide_id
        })

    return metadata


# === Main Pipeline ===
def run_pipeline_on_wsi(wsi_path, out_dir):
    slide_id = os.path.splitext(os.path.basename(wsi_path))[0]

    # Load WSI and crop thumbnail at lowest level
    reader = WSIReader.open(wsi_path)
    lowres_level = reader.info.level_count - 1
    full_dims = reader.info.level_dimensions[0]
    bounds = (0, 0, full_dims[0], full_dims[1])
    thumbnail, _, downsample = crop_bounding_box_from_wsi(wsi_path, bounds, lowres_level=lowres_level)

    # Generate full-resolution tissue mask
    mask_array = generate_tissue_mask_otsu_tiatoolbox(thumbnail, bounds, full_dims, downsample)

    # Tile and save
    metadata = tile_and_save_patches(reader, mask_array, out_dir, slide_id,
                                     patch_size=patch_size,
                                     stride=stride,
                                     min_mask_ratio=min_mask_ratio)

    # Save metadata as CSV
    df_meta = pd.DataFrame(metadata)
    df_meta.to_csv(os.path.join(out_dir, f"{slide_id}_tile_metadata.csv"), index=False)
    print(f"Done with {slide_id}, saved {len(df_meta)} tiles.")

```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

def visualize_cropped_region_with_mask(region_image, mask, alpha=0.4):
    """
    Overlay the binary tissue mask on the WSI region image.
    
    Args:
        region_image (np.ndarray): Cropped region image (H, W, 3)
        mask (np.ndarray): Binary tissue mask (H, W) or (H, W, 1)
        alpha (float): Transparency of the overlay
    """
    if mask.ndim == 3:
        mask = mask.squeeze()
    if region_image.ndim == 2 or region_image.shape[2] == 1:
        region_image = cv2.cvtColor(region_image, cv2.COLOR_GRAY2RGB)

    # Create red overlay
    red_overlay = np.zeros_like(region_image)
    red_overlay[:, :, 0] = mask  # Red channel

    # Blend image and mask
    blended = cv2.addWeighted(region_image, 1.0, red_overlay, alpha, 0)

    # Plot
    plt.figure(figsize=(10, 10))
    plt.imshow(blended)
    plt.axis("off")
    plt.title("Cropped Region with Tissue Mask Overlay")
    plt.show()
```

```python
def crop_region_and_mask(wsi_path, bounds, level=0):
    reader = WSIReader.open(wsi_path)
    region = reader.read_bounds(bounds, resolution=level, units="level")

    # Get a thumbnail for tissue detection at low resolution
    lowres_level = reader.info.level_count - 1
    thumbnail = reader.read_bounds(bounds, resolution=lowres_level, units="level")
    downsample = reader.info.level_downsamples[lowres_level]

    # Generate tissue mask
    masker = OtsuTissueMasker()
    masker.fit([thumbnail])
    masks = masker.transform([thumbnail])
    lowres_mask = masks[0].astype("uint8") * 255

    # Resize mask to match the region size at level 0
    w, h = bounds[2] - bounds[0], bounds[3] - bounds[1]
    fullres_mask = cv2.resize(lowres_mask, (w, h), interpolation=cv2.INTER_NEAREST)

    return reader, region, fullres_mask, bounds
```

```python
import time
import cv2
import numpy as np
import matplotlib.pyplot as plt
from tiatoolbox.wsicore.wsireader import WSIReader
from tiatoolbox.tools.tissuemask import MorphologicalMasker  # or OtsuTissueMasker

# Step 1: Open WSI
reader = WSIReader.open(wsi_path)

# Step 2: Define region
region_width = x_max - x_min
region_height = y_max - y_min
bounds = (x_min, y_min, x_max, y_max)

# Step 3: Read full-resolution region (this is what you're tiling)
start = time.time()
region_img = reader.read_rect(location=(x_min, y_min),
                              size=(region_width, region_height),
                              resolution=0,
                              units="level")
end = time.time()
print(f"Read region in {end - start:.2f} sec, shape: {region_img.shape}")

# Step 4: Read low-res version for mask generation
lowres_level = min(reader.info.level_count - 1, 2)
region_thumb = reader.read_rect(location=(x_min, y_min),
                                 size=(region_width, region_height),
                                 resolution=lowres_level,
                                 units="level")

# Step 5: Generate tissue mask
masker = MorphologicalMasker()  # or OtsuTissueMasker()
masker.fit([region_thumb])
mask_lowres = masker.transform([region_thumb])[0].astype(np.uint8) * 255

# Step 6: Resize mask to full resolution
mask_fullres = cv2.resize(mask_lowres, (region_width, region_height), interpolation=cv2.INTER_NEAREST)

# Step 7: Visualize overlay (mask on region)
def visualize_mask_overlay(img, mask, alpha=0.4):
    """Overlay tissue mask on image for visualization."""
    overlay = np.zeros_like(img)
    overlay[..., 0] = mask  # Red channel for tissue mask
    blended = cv2.addWeighted(img, 1.0, overlay, alpha, 0)
    plt.figure(figsize=(10, 10))
    plt.imshow(blended)
    plt.axis("off")
    plt.title("Tissue Mask Overlay on Region")
    plt.show()

visualize_mask_overlay(region_img, mask_fullres)
```
