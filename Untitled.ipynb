```python
import numpy as np
import cv2
from tiatoolbox.wsicore.wsireader import WSIReader
from skimage.color import rgb2hed
from skimage.filters import gaussian, threshold_otsu
from skimage.morphology import remove_small_objects, remove_small_holes
from skimage.measure import label


def get_tissue_mask_basic(rgb_image, deconvolve_first=True, sigma=1.5, min_size=300):
    """Your tissue detection function."""
    if deconvolve_first:
        hed = rgb2hed(rgb_image)
        hema = -hed[:, :, 0]
        hema = (hema - np.min(hema)) / (np.max(hema) - np.min(hema))
        gray_image = hema
    else:
        gray_image = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2GRAY) / 255.0

    smooth = gaussian(gray_image, sigma=sigma)
    thresh = threshold_otsu(smooth)
    mask = smooth > thresh
    mask = remove_small_objects(mask, min_size=min_size)
    mask = remove_small_holes(mask, area_threshold=min_size)
    labeled = label(mask)
    return labeled, mask.astype(np.uint8)


def read_safe_region_from_wsi(wsi_path, bounds, max_pixels=4_000_000, return_mask=True):
    """
    Safely read a bounding box region from a WSI, using TIAToolbox.
    
    Parameters:
        wsi_path (str): Path to the WSI (.tif, .svs, etc.)
        bounds (list): [x_min, y_min, x_max, y_max] in level-0 coordinates
        max_pixels (int): Max pixel count to allow (default: 4MP)
        return_mask (bool): Whether to run tissue detection on the region

    Returns:
        dict with:
            - 'region': RGB image (numpy array)
            - 'labeled': Labeled mask (optional)
            - 'mask': Binary mask (optional)
            - 'scale_info': resolution level used
    """
    reader = WSIReader.open(wsi_path)

    width = bounds[2] - bounds[0]
    height = bounds[3] - bounds[1]
    area = width * height

    if area <= max_pixels:
        resolution = 0
    else:
        # Increase resolution level until area fits under max_pixels
        level = 1
        while level < reader.level_count:
            new_w, new_h = reader.convert_bounds(bounds, in_unit="level", in_resolution=0, out_resolution=level)
            if new_w[2] - new_w[0] <= 0 or new_h[2] - new_h[0] <= 0:
                break
            if (new_w[2] - new_w[0]) * (new_h[2] - new_h[0]) <= max_pixels:
                resolution = level
                break
            level += 1
        else:
            resolution = reader.level_count - 1

    region = reader.read_bounds(bounds, resolution=resolution, units="level")

    result = {"region": region, "scale_info": resolution}

    if return_mask:
        labeled, mask = get_tissue_mask_basic(region)
        result.update({"labeled": labeled, "mask": mask})

    return result
```

```python
import time
import numpy as np
from tiatoolbox.wsicore.wsireader import WSIReader
import tifffile
import os

# ----------------------------
# CONFIGURATION
# ----------------------------
  # Input WSI in .tif format
output_path = "extracted_region_fullres.tif"  # Output TIFF path
bounds = [8209, 200, 59972, 34836]  # [x_min, y_min, x_max, y_max]

# ----------------------------
# START PROCESS
# ----------------------------
start_time = time.time()
print(f" Opening WSI: {os.path.basename(wsi_path)}")

reader = WSIReader.open(wsi_path)

print(" Extracting region at full resolution (level 0)...")
img = reader.read_bounds(bounds, resolution=0, units="level")
img_np = np.array(img)

print(f" Extracted region shape: {img_np.shape}")

# ----------------------------
# SAVE AS TIFF
# ----------------------------
print(f" Saving extracted region to {output_path}...")

tifffile.imwrite(
    output_path,
    img_np,
    bigtiff=True,                    # Handles large regions >4GB
    tile=(512, 512),                # Enables faster writing/viewing
    compression='jpeg',             # Use 'jpeg', 'lzw', or None
    photometric='rgb'               # Needed for color images
)

end_time = time.time()
print(f" Done! Region saved to {output_path} in {end_time - start_time:.2f} seconds")
```
